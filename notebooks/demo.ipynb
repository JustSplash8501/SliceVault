# SpliceVault Demo

This notebook demonstrates PSI computation, signature construction, and classifier training on a small GTEx-style subset (placeholder).

```python
import pandas as pd
from splicevault.signatures import build_signature_matrix
from splicevault.classify import train_classifier, predict_samples
```

```python
psi = pd.DataFrame({
    'S1': [0.9, 0.1, 0.3],
    'S2': [0.8, 0.2, 0.4],
    'S3': [0.2, 0.8, 0.7],
    'S4': [0.1, 0.9, 0.6],
}, index=['EVT1', 'EVT2', 'EVT3'])
labels = pd.Series({'S1': 'brain', 'S2': 'brain', 'S3': 'lung', 'S4': 'lung'})
signature, stats = build_signature_matrix(psi, labels, top_n=2)
clf = train_classifier(signature, labels, model_type='logreg')
pred, drivers = predict_samples(clf, signature)
pred
```
